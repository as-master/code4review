## Описание проекта

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

## Предобработка данных

```python
# Приведем название столбцов к нижнему регистру
df.columns = df.columns.str.lower()
df.head(1)
```

```python
#Найдем общее количество дубликатов
print('количество дубликатов:',df.duplicated().sum())
```

```python
#суммарное количество пропусков
print('количество пропущенных значений:', df.isnull().sum(), sep='\n')
```

Два пропуска в названии игр смело удалим.
<p>Пропуски по году издание найдены только в 269 строках, однако это составляет 1.6 % данных, которые выглядят вполне адекватно относительно других метрик. Не будем удалять эти данные. Целочисленный формат данных в большей степени соотвествует году релиза - изменим тип данных.
<p>Пропуски касающиеся отзывов критиков и пользователей, а так же рейтинга оставим без изменений. Так как заполнить их какой-либо статистической величиной нельзя, слишком индивидуальные значения, удалить так же нельзя - слишком много значений.
<p> В оценке пользователей встречается аббривиатура TBD, которая является сокращением для To Be Determined, то есть "Будет определено", логично эту величину заменить на отсутствующее значение nan.
<p> Пропуски в возрастном рейтинге обозначим категорией `undefined`

```python
# Преобразуем оценки пользователей в числовой вид, аббривиатура TBD представим логично ей соответствующим nan
df['user_score'] = pd.to_numeric(df['user_score'], errors = 'coerce').astype('float64')
```

```python
# Оценим представленные варианты возрастных рейтингов на наличие "не типичных" обозначений
df['rating'].unique()
```

```python
# Проверим названия платформ в избежание 2 и более имен для 1 платформы
df['platform'].unique()
```

```python
# Пропуски в возрастном рейтинге обозначим категорией `undefined`
df['rating'] = df['rating'].fillna('undefined')
```

```python
# Удалим пропущенные названия игр
df.dropna(subset = ['name'], inplace = True)
```

```python
# Изменит тип данных на целочисленное
df['year_of_release'] = df['year_of_release'].astype('Int64')
```

Добавим столбец с суммарным количеством проданых копий по всем регионам мира.

```python
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
df.info()
```

### Количество проданных копий по различным платформам

Посмотрим, как менялись продажи по платформам. Выберим платформы с наибольшими суммарными продажами и построим распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?

```python
platform_grouped = df.pivot_table(
            index = 'platform', 
            aggfunc = {'name':'count', 'total_sales':'sum', 'year_of_release':['min','max']})\
            .reset_index()\
            .rename(columns = {'name': 'games_qty'})
print('размер таблицы:', platform_grouped.shape)

# Избавимся от двухстрочных названий столбцов
platform_grouped.columns = ['platform','games_qty','total_sales','last_year','first_year']
#  Отсортируем по продажам копий игр
platform_grouped = platform_grouped.sort_values(by = 'total_sales', ascending = False).reset_index(drop=True)
#  Созданим столбец соссроком существования платформ
platform_grouped['platform_life'] = platform_grouped['last_year'] - platform_grouped['first_year']

platform_grouped[2:5]
```

Для определения среднего срока жизни платформы нужно избавиться от выбросов в виде платформ однодневок, платформ с очень низкими продажами, границу выберем в 15 млн. копий, т.к. ниже выбранной величины резкий спад в продажах, практически на порядок. Так же я бы убрал платформу PC, так как это во-первых явный выброс, а во вторых это не совсем "платформа". Платформа DS, так же привлекла внимание слишком большим сроком жизни, возможно где-то закралась ошибка, так как официально на Wiki платформа стартовала в 2004 году, по всей видимости на нее были просто импортированы старые игры. Платформы на начале своего жизненного цикла тоже не стоит учитывать в расчете среднего значения, за точку отсчета выберем 2012 год так как в этом и последующем году выпущены нексгены крайне популярных платформ таких как Nintendo Wii, PS3, X360.

```python
lifetime = platform_grouped.loc[
    (platform_grouped.loc[:,'total_sales'] > 15) & 
    (platform_grouped.loc[:,'first_year'] < 2012) & 
    (platform_grouped.loc[:,'platform_life'] < 28)]\
    ['platform_life'].mean().round(1)
print('средней срок сущетвования платформы',lifetime, 'года')
```

<p><b> Вывод </b> </p>

В среднем платформа существует около 8.4 года, однако платформы, находящиеся в топе и имеющие львиную долю продаж демонстрируют больший срок, который в последствии наверняка еще увеличится,  так как  они еще являются действующими. Ну и PC вне конкуренции по сроку жизни, как и говорилось выше, РС только грубо можно назвать платформой, тем более одной, так как за весь этот срок PC претерпел ряд изменений. 

### Актуальные платформы периода 2008 - 2016.

На основании вывода в п.3.1 выберем актуальный для анализа падения рынка игр с 2008 по 2016 года и возможных тенденций и прогнозов на 2017 год. Обосновано считаем, что если на платформе не запускаются новые игры на конец указанного периода, то перспективной ее считать нельзя. 

```python
# Актуальные платформы df = df[(df['year_of_release'] >= 2008) & (df["СубконтоКт1"] == 71.01)]
actual_platform = df.query('year_of_release >= 2008').pivot_table(  
            index = ['platform', 'year_of_release'], 
            aggfunc = {'name':'count', 'total_sales':'sum','year_of_release':['min','max']})\
            .reset_index()
#  Избавимся от двухстрочных названий столбцов
actual_platform.columns = ['platform', 'year', 'games_qty','total_sales','last_year','first_year']
 
print('размер таблицы:', actual_platform.shape)
actual_platform.head(3)
```

```python
# Сформируем список актуальных платформ, с новыми играми за 2016,
# без Wii, которая имеет в 2016 всего 1 игру
actual_platform_list = actual_platform[
            (actual_platform['last_year'] == 2016) & 
            (actual_platform['platform'] != 'Wii')]\
            ['platform'].tolist()

print('Список актуальных платформ:', actual_platform_list)
```

```python
# Оставим только платформы перечисленные в списке актуальных платформ
actual_platform = actual_platform.loc[actual_platform['platform'].isin(actual_platform_list)]
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,10), dpi= 80)
sns.lineplot(data = actual_platform, x = 'year', y = 'total_sales', hue = 'platform')

plt.title("Динамика изменения продаж игр на различных плаформах", fontsize=14)
plt.ylabel('Число проданных копий, млн.', fontsize=12)
plt.xlabel('Год', fontsize=12)
```

В условиях, когда все продажи так или иначе падают, нужно разобраться, а как меняется пропорциональная составляющая количества продаж копий на различных платформах. 

```python
# Оставим в исходном датафрейме только данные для актуальных на 2016 год платформ
actual_df = df.loc[(df['platform'].isin(actual_platform_list)) & (df['year_of_release'] >= 2008)]
```

```python
# Посчитаем отношение продаж игр на различных плаформах к общему числу продаж.
tmp = actual_df.pivot_table(index='year_of_release',columns='platform',values='total_sales',aggfunc=sum)
sale_ratio = tmp.divide(tmp.sum(axis=1),axis=0)

ax = sale_ratio.plot(kind='area',stacked='true', figsize=(15,7))

ax.legend(loc='upper left')
ax.set_ylim(0,1)
plt.title("Отношение продаж игр на различных плаформах", fontsize=14)
plt.ylabel('Отношение продаж', fontsize=12)
plt.xlabel('Год', fontsize=12)

plt.show()
```

<p><b> Вывод </b> </p>

Последний год продажи на всех платформах стремительно падают, пробивая очередное "дно". До этого лишь 2 две платформы PS4 и XOne смогли продемонстрировать рост и то связанный с их недавним стартом. В целом можно выделить 3 лидера: PS4,  XOne 3DS. Первый важный момент: При общем спаде продаж PS4 и 3DS удалось увеличить свое присутствие на рынке, т.е они падали не так быстро как все остальные. ### Пользовательские рейтинги платформ Xbox One и PC

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые, данное утверждение примем за нулевую гипотезу.  Чтобы проверить эту гипотезу применим метод scipy.stats.ttest_ind.

```python
# Посмотрим на статистические характеристики
df_tmp = pd.DataFrame()
df_tmp["PC"] = (actual_df
 .query('platform == "PC"')
 ['user_score']
 .describe())
df_tmp["XOne"] = (actual_df
 .query('platform == "XOne"')
 ['user_score']
 .describe())
df_tmp
```

Отметим, что дисперсии близки, так же как и средние значения.

Размер выборки меньше 2 сотен, а следовательно возможность случайных отклонений компенсировать друг друга не столь велика, значит рекомендуется повысить уровень значимости. Примем значение α равной 0.05.

При проверки гипотез возможны два варианта ошибок: Ошибка первого рода - отвергнуть верную нулевую гипотезу с вероятность α; Ошибка второго рода - принять неверную нулевую гипотезу с вероятность β. Величины α и β обратно взаимосвязаны и сделать обе ошибки равыми нулю невозможно. Поэтому нужно найти баланс, широко рекомендуется использовать величину α равной 0.05. Однако следует учитывать размер выборки, а в первой гипотезе она мала, следовательно более вероятно получить случайный результат можно взять большую вероятность ошибки первого рода α равной 0.1. Хотя теоритичесик можно увеличь вплоть до 0.2 без изменения результата. Увеличивая уровень значимости, увеличивается шансы отвергнуть нулевую гипотезу.

```python
# Лист пользовательских рейтингов PC 
sample_1 = actual_df.query('platform == "PC"')['user_score'].dropna().tolist() 

print('Размер выборки  рейтингов платформы PC:', len(sample_1))
```

```python
# Лист пользовательских рейтингов платформы Xbox One 
# Размеры выборок должны быть одинаковые для выбранного метода
sample_2 = actual_df.\
                query('platform == "XOne"')['user_score'].dropna().\
                sample(n = 155, random_state = 20).\
                tolist()

print('Размер выборки  рейтингов платформы Xbox One:', len(sample_2))
```

```python
alpha = 0.05  # критический уровень статистической значимости

results = st.ttest_ind(sample_1, sample_2, equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

<p><b> Вывод </b> </p>

Отвергнуть нулевая гипотеза о равенстве средних пользовательских рейтингов двух платформ не удалось, следовательно можно говорить от том, что Xbox One и PC имеют схожие отзывы пользователей а если различия и имеются, то не статистически значимые.

```python
# Посмотрим на статистические характеристики
df_tmp = pd.DataFrame()
df_tmp["Action"] = (actual_df
 .query('genre == "Action"')
 ['user_score']
 .describe())
df_tmp["Sports"] = (actual_df
 .query('genre == "Sports"')
 ['user_score']
 .describe())
df_tmp
```

Размер выборки, а значит уровень значимости те же. Примем значение α равной 0.05.
