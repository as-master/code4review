### Описание данных

<b>Технологический процесс </b>  
Rougher feed — исходное сырье  
Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant  
Xanthate **— ксантогенат (промотер, или активатор флотации);  
Sulphate — сульфат (на данном производстве сульфид натрия);  
Depressant — депрессант (силикат натрия).  
Rougher process (англ. «грубый процесс») — флотация  
Rougher tails — отвальные хвосты  
Float banks — флотационная установка  
Cleaner process — очистка  
Rougher Au — черновой концентрат золота  
Final Au — финальный концентрат золота  
<b> Параметры этапов </b>  
air amount — объём воздуха  
fluid levels — уровень жидкости  
feed size — размер гранул сырья  
feed rate — скорость подачи  

<b> Наименование признаков   </b>  
Наименование признаков должно быть такое:  

    [этап].[тип_параметра].[название_параметра]  
    
Пример: rougher.input.feed_ag  
<b> Возможные значения для блока [этап]:  </b>   
rougher — флотация  
primary_cleaner — первичная очистка  
secondary_cleaner — вторичная очистка  
final — финальные характеристики  
<b> Возможные значения для блока [тип_параметра]:  </b>   
input — параметры сырья  
output — параметры продукта  
state — параметры, характеризующие текущее состояние этапа  
calculation — расчётные характеристики

#### Замена нулевых значений

#### Замена пропусков медианным значением

Видно что диапозон изменения некоторых признаков весьма мал, при этом значения за рамками боксплота считаем выбросами, поэтому будет справедливо заменить пропуски в этих признаках на медианное значение, так как это мало изменит общую картину распределений.  

```python
# Список названий признаков
feature_list = list(stat_params)
# Пороговое значение размаха 1 и 3 квартилей относительно медианного значения
threshold_range = 0.3

small_quartiles_range_list = []

for i in range(56):
    try:
        range_ratio = abs((stat_params.loc['75%', feature_list[i]] - stat_params.loc['25%', feature_list[i]])/stat_params.loc['50%', feature_list[i]])
        if range_ratio <= threshold_range:
            small_quartiles_range_list.append(feature_list[i])
    except:
        print('Не обработан', feature_list[i])
```

```python
small_quartiles_range_list
```

Итак, имеем набор признаков с  малым размахом изменений значений относительно медианного значения. В данных признаках заменим пропуски медианным значением

```python
for i in range(len(small_quartiles_range_list)):
# Заполним пропуски в тренировочной выборке
    train_df[small_quartiles_range_list[i]] = train_df[small_quartiles_range_list[i]].\
    fillna(train_df[small_quartiles_range_list[i]].median())
# Заполним пропуски в тестовой выборке    
    try:
            test_df[small_quartiles_range_list[i]] = test_df[small_quartiles_range_list[i]].\
            fillna(train_df[small_quartiles_range_list[i]].median())
    except:
        print(small_quartiles_range_list[i], 'Данный признак не входит в тестовую выборку')
```

```python
# суммарное количество пропусков
test_df.isnull().sum()
```

#### Фильтрация выбросов

В предыдущем пункте на изображения боксплотов видно большое количество выбросов. Что может быть объяснено неисправностью измеряющих приборов, либо наличию шумов в сигналах. 

```python
stat_params
```

```python
for i in range(len(train_df.columns)):
    # Границы величин выбросов
    iqr = stat_params.loc['75%', feat_list[i]] - stat_params.loc['25%', feat_list[i]]
    low = stat_params.loc['25%', feat_list[i]] - 3 * iqr
    high = stat_params.loc['25%', feat_list[i]] + 3 * iqr
    # Приведем выбросы к пропускам
    train_df.loc[train_df[feat_list[i]] < low, feat_list[i]] = np.nan
    train_df.loc[train_df[feat_list[i]] > high, feat_list[i]] = np.nan
```

```python
for i in range(len(test_df.columns)):
    # Границы величин выбросов
    iqr = stat_params.loc['75%', feat_list[i]] - stat_params.loc['25%', feat_list[i]]
    low = stat_params.loc['25%', feat_list[i]] - 3 * iqr
    high = stat_params.loc['25%', feat_list[i]] + 3 * iqr
    # Приведем выбросы к пропускам
    test_df.loc[test_df[list(test_df)[i]] < low, list(test_df)[i]] = np.nan
    test_df.loc[test_df[list(test_df)[i]] > high, list(test_df)[i]] = np.nan
```
